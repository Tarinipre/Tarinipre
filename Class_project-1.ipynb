#PDF AND CDF PLOTS

The above plot will be help us understand the data set features towards classfication and informed interpretation.


The figures obtained below is what was used to plot the PDF and CDF.

```python
Data1_Died = Data1.loc[Data1['survival_status'] == 0];
Data1_Survived = Data1.loc[Data1['survival_status'] == 1];
```

```python
counts, bin_edges = np.histogram(Data1_Died["age"], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)


counts, bin_edges = np.histogram(Data1_Died['axil_nodes'], bins=20, 
                                 density = True)
pdf = counts/(sum(counts))
plt.plot(bin_edges[1:],pdf);


plt.show();
```

From the above graph, the orange line represents the Cumulative Distribution Function (CDF), while the blue line represents the Probability Density Function(PDF) for the "age"; and Axil nodes is shown with the green line.

The PDF provides more insight into the relationship between Caner patients that did not survive over 5 years of their surgery, using age and number of axil nodes.

A careful study of the PDF and CDF plots also tells us that patients above 50 have had  lesser chances of surviving 5 years after their surgery. Furthermore, the number of axil nodes found in a cancer patient could cause a higher probability of fatality case for such category of patients.

```python
#Plot CDF of age

counts, bin_edges = np.histogram(Data1_Survived['age'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)

#compute CDF
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)

counts, bin_edges = np.histogram(Data1_Survived['axil_nodes'], bins=20, 
                                 density = True)
pdf = counts/(sum(counts))
plt.plot(bin_edges[1:],pdf);



plt.show();
```

With comparison to the previous graph, the PDF of the above graph is blue while CDF is orange, and axil nodes, green.

From the graph, we can tell that patients below 50 had a higher probability of surviving 5 years after their operation, likewise those with lesser numbers of axil nodes.

##**Classification**

```python
Data1.groupby([Data1.age>=51, 'survival_status']).count()
```

```python
# Importing the dataset
X = Data1.iloc[:, [0, 2]].values
y = Data1.iloc[:, 3].values
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.35, random_state = 0)
```

```python
X_test
```

```python
X_train
```

```python
# Fitting Logistic Regression to the Training set
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
y_train
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print (cm)
```
